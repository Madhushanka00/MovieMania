```python
import pandas as pd
import numpy as np
```

```python
df_links = pd.read_csv('links.csv')
df_links.head()
```

```python
df_movies = pd.read_csv('movies.csv')
df_movies.head()
```

```python
split_genres = df_movies['genres'].str.split('|', expand=True).stack()
print(split_genres)
```

```python
df_dummies = pd.get_dummies(split_genres)
print(df_dummies)
```

```python
df_encoded = df_dummies.groupby(level=0).sum()
print(df_encoded)
```

```python
df_encoded.head()
```

```python
df_encoded = pd.concat([df_movies['movieId'], df_encoded], axis=1)
```

```python
df_encoded.head()
```

```python
genre_counts = {}

for genre in df_dummies.columns[1:]:
    genre_counts[genre] = df_encoded[df_encoded[genre] == 1].shape[0]

for genre, count in genre_counts.items():
    print(f"Number of {genre} movies: {count}")
```
