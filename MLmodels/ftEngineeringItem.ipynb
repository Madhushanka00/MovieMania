```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_ratings = pd.read_csv('ratings.csv')
df_ratings.drop(columns=['timestamp'], inplace=True)
df_ratings.head()
```

```python
average_rating_per_movie = df_ratings.groupby('movieId')['rating'].mean().reset_index()

# Rename the columns for clarity
average_rating_per_movie.columns = ['movieId', 'avg_rating']

# Display the new DataFrame
average_rating_per_movie.head()
```

```python
df_movies = pd.read_csv('movies.csv')
df_movies.drop(columns=['genres'], inplace=True)
df_movies.head()
```

```python
# Extract the year from the 'title' column
df_movies['year'] = df_movies['title'].str.extract(r'\((\d{4})\)')
df_movies.drop(columns=['title'], inplace=True)
df_movies.head()
```

```python
df_encoded = pd.read_csv('Movies_encoded.csv')
df_encoded.head()
```

```python
combined_movieDf = pd.merge(df_movies,average_rating_per_movie, on='movieId')
combined_movieDf = pd.merge(combined_movieDf,df_encoded, on='movieId')
combined_movieDf.head()
```

```python
df_ratings.drop(columns=['userId'], inplace=True)
df_ratings.drop(columns=['rating'], inplace=True)
df_ratings.head()
```

```python
item_train = pd.merge(df_ratings,combined_movieDf, on='movieId')
item_train.head()
```

```python
item_train.to_csv('item_train.csv', index=False)
print('item_train.csv saved')
```
