motivation of this notebook is  create a new feature per user, as avarage rating for each genres

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
df_user = pd.read_csv('ratings.csv')
df_user = df_user.drop(columns=['timestamp'])
df_movies = pd.read_csv("Movies_encoded.csv")
df_user.head()
```

```python
df_movies.head()
```

```python
user_1_data = df_user[df_user['userId'] == 1]
print(user_1_data)
```

```python
combined_df = pd.merge(df_user, df_movies, on='movieId')
combined_df.head()
```

```python
genras = ['Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'IMAX', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

for column in genras:
    combined_df[column] = combined_df.apply(lambda row: row['rating'] if row[column] == 1 else row[column], axis=1)

combined_df = combined_df.drop(columns=['rating'])
```

```python
combined_df.head(1000)
```

```python
user_ratings = combined_df[combined_df['userId'] == 100]
user_ratings.head()
```

```python
# Filter the DataFrame to include only rows where the 'Action' column has non-zero values
action_ratings = user_ratings[user_ratings['Action'] != 0]

# Calculate the average rating for the 'Action' genre
average_action_rating = round(action_ratings['Action'].mean(),2)

print(f"Average rating for 'Action' genre: {average_action_rating}")
```

```python
for column in genras:
    genre_ratings = user_ratings[user_ratings[column] != 0]
    average_rating = round(genre_ratings[column].mean(),2)
    print(f"Average rating for '{column}' genre: {average_rating}")
```

```python
def calculateRatings(userid):
    avg_array =[]
    user_ratings = combined_df[combined_df['userId'] == userid]
    genras = ['Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'IMAX', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

    for column in genras:
        genre_ratings = user_ratings[user_ratings[column] != 0]
        average_rating = round(genre_ratings[column].mean(),2)
        avg_array.append(average_rating)
    return avg_array
```

```python
# Initialize an empty DataFrame
all_user_ratings = pd.DataFrame()

# Loop through each user ID and calculate ratings
for userId_item in range(1, 611):
    ratings = calculateRatings(userId_item)
    data = {'user_id': [userId_item] , 'Action':[ratings[0]], 'Adventure':[ratings[1]], 'Animation':[ratings[2]], 'Children':[ratings[3]], 'Comedy':[ratings[4]], 'Crime':[ratings[5]], 'Documentary':[ratings[6]], 'Drama':[ratings[7]], 'Fantasy':[ratings[8]], 'Film-Noir':[ratings[9]], 'Horror':[ratings[10]], 'IMAX':[ratings[11]], 'Musical':[ratings[12]], 'Mystery':[ratings[13]], 'Romance':[ratings[14]], 'Sci-Fi':[ratings[15]], 'Thriller':[ratings[16]], 'War':[ratings[17]], 'Western':[ratings[18]]}
    df = pd.DataFrame(data)
    all_user_ratings = pd.concat([all_user_ratings, df], ignore_index=True)


# Save the pivoted DataFrame to a CSV file
all_user_ratings.to_csv('all_user_ratings.csv', index=False)

print("Data saved to all_user_ratings.csv")
```
